```python
import pandas as pd
from pathlib import Path

def quick_analyse_dataset(file_path, dataset_name=None, source=None, domain=None, coverage_notes=None):
    dataset_name = dataset_name or Path(file_path).stem
    ext = Path(file_path).suffix.lower()

    # Load
    if ext == ".csv":
        df = pd.read_csv(file_path)
    elif ext in [".xlsx", ".xls"]:
        df = pd.read_excel(file_path)
    else:
        raise ValueError(f"Unsupported file format: {ext}")

    # Basic info
    summary = {
        "Dataset Name": dataset_name,
        "Source": source,
        "Domain": domain,
        "Rows": df.shape[0],
        "Columns": df.shape[1],
        "Coverage": coverage_notes,
        "Column Names": list(df.columns),
        "Column Types": df.dtypes.astype(str).to_dict(),
        "Null Counts": df.isnull().sum().to_dict(),
        "Example Rows": df.head(3).to_dict(orient="records")
    }

    return summary, df


# Example usage
dataset_summary, df = quick_analyse_dataset(
    file_path="State-wise-Crop-Yield-data.csv",
    dataset_name="Example Weather Data",
    source="https://example.com/dataset",
    domain="Weather",
    coverage_notes="Covers all districts of Maharashtra from 2015–2023"
)

import pprint
pprint.pprint(dataset_summary)
```
